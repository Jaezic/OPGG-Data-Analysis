```python
# Load Dataset CSV File
import pandas as pd 

dataset = pd.read_csv('opgg.csv')

# Display the first 5 rows of the dataset
print(dataset.head())

# Display the Statistics of the dataset
print(dataset.describe())

# Visualize the statistics of the dataset
import matplotlib.pyplot as plt
dataset.hist(layout=(3,3), figsize=(10,10), bins=20)
plt.show()
```

```python
# Preprocessing the dataset
# Most Champ Columns are Categorical and List, so we will convert them to numerical values using OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

# Find out the All the Unique Values in the Champ Columns
champs = ()
for most_champs in dataset['Most Champ']:
    for champ in eval(most_champs):
        if champ not in champs:
            champs += (champ,)

# Add the Unique Values to the DataFrame as Columns
for champ in champs:
    dataset[champ] = 0

# Set the Champ Columns to 1 if the Champ is in the Most Champ List
for i, most_champs in enumerate(dataset['Most Champ']):
    for champ in eval(most_champs):
        dataset.at[i, champ] = 1
```

```python
# Display the first 5 rows of the dataset
print(dataset.head())

# Display the Statistics of the dataset
print(dataset.describe())
```

```python
#
# Most Used Champ Columns Visualization
import matplotlib.pyplot as plt
import seaborn as sn

# Count the number of times each Champ is used
champ_count = {}
for champ in champs:
    champ_count[champ] = dataset[champ].sum()

# Sort the Champ Count
champ_count = dict(sorted(champ_count.items(), key=lambda item: item[1], reverse=True))

# Plot the Champ Count

plt.figure(figsize=(20,10))
plt.bar(champ_count.keys(), champ_count.values())
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation between LP and Win Ratio

# Correlation Matrix
corrMatrix = dataset[['LP', 'Win Ratio']].corr()

# Heatmap
sn.heatmap(corrMatrix, annot=True)
plt.show()

# LP ÏôÄ ÏäπÎ•†Ïùò Scatter Plot ÌôïÏù∏
plt.scatter(dataset['LP'], dataset['Win Ratio'])
plt.xlabel('LP')
plt.ylabel('Win Ratio')
plt.show()
```

```python
# Correlation between Most Used Champ and Win Ratio
# Correlation Matrix
corrMatrix = dataset[['Win Ratio', 'Lucian', 'Senna', 'Graves']].corr()

# Heatmap
sn.heatmap(corrMatrix, annot=True)
plt.show()

# Correlation between Less Used Champ and Win Ratio
# Correlation Matrix
corrMatrix = dataset[['Win Ratio', 'Shyvana', 'Yorick', 'Skarner']].corr()

# Heatmap
sn.heatmap(corrMatrix, annot=True)
plt.show()
```
